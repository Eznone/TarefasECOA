Importing Dependencies

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

Gathering the Image 
Getting dimensions of the Image

```python
original = cv2.imread("../images/marioGame.png", cv2.IMREAD_GRAYSCALE)
assert original is not None, "Comparee Image doesn't exist"
image = original.copy()
template = cv2.imread("../images/coin.png", cv2.IMREAD_GRAYSCALE)
assert template is not None, "Template Image doesn't exist"
w, h = template.shape[::-1]

plt.imshow(image)
plt.title("Comparee")
plt.show()
plt.imshow(template)
plt.title("Comparer")
plt.show()
```

TM_CCOEFF method

```python
img = image.copy()
method = getattr(cv2, "TM_CCOEFF")

# Threshold Application
res = cv2.matchTemplate(img, template, method)
min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(res)
top_left = min_loc
bottom_right = (top_left[0] + w, top_left[1] + h)
cv2.rectangle(img, top_left, bottom_right, 255, 2)

plt.subplot(121),plt.imshow(res,cmap = 'gray')
plt.title('Matching Result'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(img,cmap = 'gray')
plt.title('Detected Point'), plt.xticks([]), plt.yticks([])
plt.suptitle("TM_CCOEFF")
plt.show()
```
